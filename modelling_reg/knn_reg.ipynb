```python
import os

import subprocess
import time

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import numpy as np
import pandas as pd

from utils import load_data

#work_dir='/nagyvinyok/adat84/sotejedlik/ribli/mhc'
work_dir='/mnt/Data1/ribli/mhc/'
os.chdir(work_dir)
```

```python
x,y=load_data(hla_representation='one-hot',seq_representation='one-hot')
```

```python
from sklearn.cross_validation import cross_val_predict
from sklearn.neighbors import KNeighborsRegressor

model=KNeighborsRegressor(n_neighbors=5,n_jobs=4)
# note this takes some tens of minutes
y_pred=cross_val_predict(model,x,y,cv=2)
```

```python
plt.figure(figsize=(9,9))
plt.plot(y,y_pred,'.',alpha=0.4)
plt.plot(y,y,lw=3,alpha=0.5)
plt.xlim(xmin=-0.5)
plt.ylim(-0.5,8)
from scipy.stats import pearsonr
print pearsonr(y,y_pred)
```
